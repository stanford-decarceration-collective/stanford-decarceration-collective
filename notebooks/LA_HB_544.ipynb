# Turning sentence_length into transitions_data

```python
sentence_length
```

```python
transitions_data = sentence_length.copy()
transitions_data = transitions_data.iloc[1:]

for i in range(len(crime_types)):
    num = '' if i == 0 else '.' + str(i)
    transitions_data[crime_types[i]] = transitions_data['Black' + num].astype(float) + transitions_data['White' + num].astype(float)
    
transitions_data = transitions_data[crime_types]

transitions_data.loc['50-50'] = transitions_data.loc['50+'] + transitions_data.loc['death'] + transitions_data.loc['life']
transitions_data = transitions_data.drop(['death', 'life', '50+'])

transitions_data
```

```python
transitions_data = transitions_data.reset_index()
```

```python
transitions_data = transitions_data.rename(columns={'index': 'compartment_duration'})
```

```python
def punk_jr(compartment_duration):
    compartment_duration = compartment_duration.split('-')
    start = float(compartment_duration[0])
    final_duration = start
    
    if len(compartment_duration) > 1:
        end = float(compartment_duration[1])
        final_duration = -start + end + 1
        
    return final_duration * 12

transitions_data_jr = transitions_data.copy()
transitions_data_jr["months_in_group"] = transitions_data_jr['compartment_duration'].apply(punk_jr)


# disallow sentences less than 6 months
transitions_data_jr.loc[0, 'compartment_duration'] = '0.5-2'
transitions_data_jr.loc[0, 'months_in_group'] = 30.0

transitions_data_jr
```

```python
transitions_data_jr.property.sum() + transitions_data_jr.violent.sum() + transitions_data_jr.drug.sum() + transitions_data_jr.others.sum()
```

```python
transitions_jr_the_second = pd.DataFrame()
crime_types = ['violent', 'property', 'drug', 'others']
prev_months = 6

for i, row in transitions_data_jr.iterrows():
    for j in crime_types:
        for month in range(int(prev_months), int(prev_months + row['months_in_group'])):
            new_row = {'compartment_duration': month, 'crime_type': j, 'total_population': row[j]/(month * row['months_in_group']),
                   'compartment': 'prison', 'outflow_to': 'release'}
            transitions_jr_the_second = transitions_jr_the_second.append(new_row, ignore_index=True)
    prev_months += row['months_in_group']
```

```python
#scale down compartment_duration to convert from sentence length to LOS.
avg_LOS = (
    transitions_jr_the_second.compartment_duration * transitions_jr_the_second.total_population
).sum()/transitions_jr_the_second.total_population.sum()

#actual LOS taken from https://www.pewtrusts.org/-/media/legacy/uploadedfiles/pcs_assets/2012/prisontimeservedlouisianapdf.pdf
LOS_scale_factor = 2.5 * 12 / avg_LOS
transitions_jr_the_second.compartment_duration *= LOS_scale_factor

#life sentences should not get shortened
transitions_jr_the_second.loc[
    transitions_jr_the_second.compartment_duration >= 50 * 12 * LOS_scale_factor, 'compartment_duration'
] = 50.0 * 12

#sentences shouldn't get shorter than six months
transitions_jr_the_second.loc[
    transitions_jr_the_second.compartment_duration < 6, 'compartment_duration'
] = 6.0

LOS_scale_factor
```

```python
reincarceration_transitions = transitions_jr_the_second.copy()
reincarceration_transitions.compartment = 'reincarceration'
transitions_jr_the_second = pd.concat([transitions_jr_the_second, reincarceration_transitions])
```

```python
subgroups = transitions_jr_the_second.crime_type.unique()
transitions_jr_the_second = transitions_jr_the_second.append(pd.DataFrame({
    'compartment_duration': [1] * len(subgroups),
    'crime_type': subgroups,
    'total_population': [1] * len(subgroups),
    'compartment': ['release_full'] * len(subgroups),
    'outflow_to': ['release_full'] * len(subgroups)
}), ignore_index = True)

transitions_jr_the_second.compartment_duration = transitions_jr_the_second.compartment_duration.astype(float)
transitions_jr_the_second.total_population = transitions_jr_the_second.total_population.astype(float)




transitions_jr_the_second 
```

```python
transitions_jr_the_second.groupby('compartment_duration').sum()['total_population'].plot(x='compartment_duration', y='total_population', ylim=[0, None])
```

```python
cumulative_absolute_recidivism = recidivism.iloc[:5,:-2].astype(float)
absolute_recidivism = pd.DataFrame().append(cumulative_absolute_recidivism.iloc[0], ignore_index=True)
for i in range(1,5):
    absolute_recidivism = absolute_recidivism.append(
        cumulative_absolute_recidivism.iloc[i] - cumulative_absolute_recidivism.iloc[i - 1],
        ignore_index=True
    )
absolute_recidivism.index = cumulative_absolute_recidivism.index
absolute_recidivism
```

```python
recidivism_by_crime_type = recidivism.iloc[10:15,:4].set_axis(recidivism.iloc[9,:4], axis=1).astype(float).rename(
    {'other': 'others'}, axis=1
)
recidivism_by_crime_type
```

```python
absolute_recidivism_by_crime_type = recidivism_by_crime_type.multiply(
    absolute_recidivism['Total Population'], axis=0
).rename({'other': 'others'}, axis=1)
absolute_recidivism_by_crime_type
```

```python
recidivism_by_history = recidivism.iloc[17:22,:4].set_axis(recidivism.iloc[16,:4], axis=1).astype(float).rename(
    {'other': 'others'}, axis=1
)
recidivism_by_history
```

```python
#WARNING: you have to recreate outflows_data each time you run this, cuz this cell messes with outflows_data

recidivism_transitions = pd.DataFrame()
total_recidivism = absolute_recidivism['Total Population'].sum()

for crime_type in recidivism_by_crime_type.columns:
    recidivism_transitions = recidivism_transitions.append(pd.DataFrame({
        'compartment': ['release'] * 6,
        'outflow_to': ['reincarceration'] * 5 + ['release_full'],
        'crime_type': crime_type,
        'compartment_duration': [12,24,36,48,60, 60],
        'total_population': list(absolute_recidivism['Total Population']) + [1 - total_recidivism]
    }), ignore_index=True)

transitions_jr_the_second = pd.concat([transitions_jr_the_second, recidivism_transitions])
outflows_data.total_population *= (1 - total_recidivism)
    
```

# Modeling

```python
# Don't worry about any warnings, as long as this doesn't fail you're in business
spark_sim = SuperSimulationFactory.build_super_simulation(
    '../state/LA/policy_configurations/HB_544_model_inputs.yaml'
)
```

```python
spark_sim.simulate_baseline(['prison', 'reincarceration'])
```

```python
max_earned_time = 7 / 20
non_earned_time_reduction = max_earned_time - LOS_scale_factor

second_incarceration_plus_fraction = recidivism_by_history.mean().iloc[1:].sum()
fourth_incarceration_plus = recidivism_by_history.mean()['4th+ offense']

# NONVIOLENT

non_violent_policy_transitions = transitions_jr_the_second[
    (transitions_jr_the_second.compartment == 'reincarceration') &
    (transitions_jr_the_second.crime_type != 'violent')
].copy()

shortened_non_violent = non_violent_policy_transitions.copy()

shortened_non_violent.loc[
    shortened_non_violent.compartment_duration != 600, 'compartment_duration'
] *= 0.65 / LOS_scale_factor

shortened_non_violent.total_population *= fourth_incarceration_plus
non_violent_policy_transitions.total_population *= 1 - fourth_incarceration_plus

non_violent_policy_transitions = pd.concat([non_violent_policy_transitions, shortened_non_violent])

# VIOLENT

violent_policy_transitions = transitions_jr_the_second[
    (transitions_jr_the_second.compartment == 'reincarceration') &
    (transitions_jr_the_second.crime_type == 'violent')
].copy()

shortened_violent = violent_policy_transitions.copy()

shortened_violent.loc[
    shortened_violent.compartment_duration != 600, 'compartment_duration'
] *= (1 - non_earned_time_reduction) / LOS_scale_factor

shortened_violent.total_population *= second_incarceration_plus_fraction
violent_policy_transitions.total_population *= 1 - second_incarceration_plus_fraction

violent_policy_transitions = pd.concat([violent_policy_transitions, shortened_violent])


def reduce_earned_time(policy_transitions):
    return partial(TransitionTable.use_alternate_transitions_data, 
                   alternate_historical_transitions=policy_transitions,
                   retroactive=True)

policy_list = []
policy_list.append(SparkPolicy(policy_fn=reduce_earned_time(violent_policy_transitions),
                                       spark_compartment='reincarceration',
                                       sub_population={'crime_type': 'violent'},
                                       policy_ts=36, #jan 2023
                                       apply_retroactive=True))
policy_list.append(SparkPolicy(policy_fn=reduce_earned_time(non_violent_policy_transitions),
                                       spark_compartment='reincarceration',
                                       sub_population={'crime_type': 'others'},
                                       policy_ts=36, #jan 2023
                                       apply_retroactive=True))
policy_list.append(SparkPolicy(policy_fn=reduce_earned_time(non_violent_policy_transitions),
                                       spark_compartment='reincarceration',
                                       sub_population={'crime_type': 'property'},
                                       policy_ts=36, #jan 2023
                                       apply_retroactive=True))
policy_list.append(SparkPolicy(policy_fn=reduce_earned_time(non_violent_policy_transitions),
                                       spark_compartment='reincarceration',
                                       sub_population={'crime_type': 'drug'},
                                       policy_ts=36, #jan 2023
                                       apply_retroactive=True))


test_results = spark_sim.simulate_policy(policy_list, 'reincarceration')
```
